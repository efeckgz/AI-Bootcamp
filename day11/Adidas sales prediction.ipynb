# Adidas sale prediction - time series

```python
import pandas as pd
from datetime import date, timedelta
import datetime
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima_model import ARIMA
import statsmodels.api as sm
import plotly.express as px
```

ARIMA: Auto regressive integration moving average

```python
df = pd.read_csv("adidas-quarterly-sales.csv")
df.head()
```

```python
px.line(df, x='Time Period', y='Revenue')
```

```python
result = seasonal_decompose(df['Revenue'], model='multiplicative', period=30)
fig = result.plot()
fig.set_size_inches(15, 10)
```

```python
pd.plotting.autocorrelation_plot(df['Revenue']);
```

```python
model = sm.tsa.statespace.SARIMAX(df['Revenue'])
results = model.fit()
results.summary()
```

```python
predictions = results.predict(len(df), len(df) + 7)
predictions
```

```python
df['Revenue'].plot();
predictions.plot();
```
