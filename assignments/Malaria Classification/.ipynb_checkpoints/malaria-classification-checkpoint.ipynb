## Preparing Data

In this stage we will read the image files into a dataframe. This dataframe will have two columns.

1. The path of the image file
2. A boolean column that represents weather the image is of infected cell or not.

```python
labels = ['Parasitized', 'Uninfected']
impath = '/kaggle/input/cell-images-for-detecting-malaria/cell_images/'
```

```python
images_df = []
labels_df = []

for label in labels:
    for imfile in os.listdir(impath + label):
        # Ignore dotfiles, if any
        if imfile[0] == '.':
            continue
        images_df.append(f'{impath}{label}/{imfile}')
        labels_df.append(label)
```

```python
df = pd.DataFrame({
    'Image': images_df,
    'Label': labels_df
})

df.head()
```

```python
df['Infected'] = df['Label'].map({
    'Parasitized': True,
    'Uninfected': False
})
df.drop('Label', axis=1, inplace=True)
df.sample(10)
```

## Show Images

I will display 3 images that are infected and 3 that are not infected. The images will be selected randomly.

```python
import matplotlib.pyplot as plt
```

### Infected images

```python
infected = df[df['Infected'] == True]
infected = infected.sample(3)
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

for i, impath in enumerate(infected['Image']):
    im = cv2.imread(impath)
    im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
    ax[i].imshow(im)
    ax[i].axis('off')

fig.suptitle('Infected cell images', fontsize=16)
plt.show()
```

### Uninfected images

```python
uninfected = df[df['Infected'] == False]
uninfected = uninfected.sample(3)
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

for i, impath in enumerate(uninfected['Image']):
    im = cv2.imread(impath)
    im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
    ax[i].imshow(im)
    ax[i].axis('off')
    
fig.suptitle('Uninfected cell images', fontsize=16)
plt.show()
```
