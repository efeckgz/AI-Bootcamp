```python
from sklearn import datasets
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
data = datasets.load_iris()
x = data.data
y = data.target
```

```python
km = KMeans(random_state=42)
visualizer = KElbowVisualizer(km, k=(2, 10))
```

```python
visualizer.fit(x)
visualizer.show()
```
