## Modelling 

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, InputLayer, Reshape, MaxPooling2D, Flatten, BatchNormalization, Dropout
from tensorflow.keras.losses import SparseCategoricalCrossentropy
```

### Normalize the pixels

Each pixel holds a (r, g, b) value that ranges from 0 to 255. We can divide the rgb values of each pixel by 255 to normalize this to 1.

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
# model = Sequential()

# # Input layer
# model.add(InputLayer(input_shape=train_shape))

# # Convolutional layers
# model.add(Conv2D(filters=32, kernel_size=(3, 3), activation='relu'))

# # Pooling layer
# model.add(MaxPooling2D(pool_size=(2, 2)))

# model.add(Flatten())
# model.add(BatchNormalization())
# model.add(Dense(10))

# # Compile
# model.compile(optimizer='adam', loss=SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
```

```python
model = Sequential()

# Input layer
model.add(InputLayer(input_shape=train_shape))

# Convolutional layers
model.add(Conv2D(filters=32, kernel_size=(3, 3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

# Flattening the layers
model.add(Flatten())

# Dense layers
model.add(Dense(256, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(10))  # No activation here since we'll use from_logits=True

# Compile
model.compile(optimizer='adam', loss=SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
```

```python
history = model.fit(train_images, train_labels, epochs=100)
```

```python
loss, accuracy = model.evaluate(test_images, test_labels)
print(f"Loss: {loss * 100:.2f}%\nAccuracy: {accuracy * 100:.2f}%")
```
