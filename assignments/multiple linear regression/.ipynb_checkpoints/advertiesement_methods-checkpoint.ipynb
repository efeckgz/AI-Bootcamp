## Modelling

```python
from sklearn.linear_model import LinearRegression as LR
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import r2_score
```

```python
lr = LR()
```

```python
x = df.drop('Sales', axis=1)
y = df['Sales']
```

```python
x_train, x_test, y_train, y_test = tts(x, y, random_state=42, test_size=.2)
```

```python
model = lr.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
```

```python
print(f'Mean squared error: {mse(y_test, prediction)}')
print(f'R2 score: {r2_score(y_test, prediction)}')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(y_test, prediction)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='black')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs. Predicted Sales')
plt.show()
```

```python
coefs = pd.DataFrame(lr.coef_, x.columns, columns=['Coefficient'])
coefs.head()
```
