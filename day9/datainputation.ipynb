## Filling empty values

### LinearRegression

```python
from sklearn.experimental import enable_iterative_imputer # We need this for iterative imputer to work.
from sklearn.impute import IterativeImputer
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
imptr = IterativeImputer(estimator=lr)
imputed = imptr.fit_transform(df)
```

```python
pd.DataFrame(imputed)
```

### KNNImputer

```python
from sklearn.impute import KNNImputer
```

```python
knn = KNNImputer()
```

```python
imptd_knn = knn.fit_transform(df)
pd.DataFrame(imptd_knn, columns=df.columns) # With KNN we need to specify columns.
```

### SimpleImputer

This is a simpler approach. I uses the provided statistical data about each column to fill in missing values. A more automatic approach to what we have been doing earlier.

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(strategy='mean')
```

```python
imptd_si = si.fit_transform(df)
pd.DataFrame(imptd_si, columns=df.columns)
```
