```python
page = requests.get('https://www.karar.com/yazarlar/zafer-acar')
soup = BS(page.content, 'html.parser')
```

```python
links = soup.find_all('article', { 'class': 'item box-shadow' })
```

```python
links
```

```python
texts = []
headers = []
for link in links:
    _link = f'https://www.karar.com{link.a.get("href")}'
    header = link.h3.text
    
    page2 = requests.get(_link)
    soup2 = BS(page2.content, 'html.parser')

    doc = soup2.select('article.article-content')[0].text

    texts.append(doc)
    headers.append(header)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=['Title', 'Article'])
df['Title'] = headers
df['Article'] = texts
```
